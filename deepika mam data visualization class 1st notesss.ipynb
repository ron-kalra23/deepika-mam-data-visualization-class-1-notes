Inside the matplotlib there is a plot name as bar plot .let's see the bar plot 

```python
x =[5,10,12,13,7]
y = [12,13,16,20,8]
plt.bar(x,y)
plt.title("Bar graph")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x = ['mango' , 'apple' , 'banana ' , 'kiwi']
y=[100 , 70,80,300]
plt.bar(x,y) # by default there is a parameter allign = 'center'
plt.title(" Total sales")
plt.xlabel('fruits')
plt.ylabel('sales')
plt.show()
```

```python
x = ['mango' , 'apple' , 'banana ' , 'kiwi']
y=[100 , 70,80,300]
plt.bar(x,y , align = 'edge') # by default there is a parameter allign = 'center'
plt.title(" Total sales")
plt.xlabel('fruits')
plt.ylabel('sales')
plt.show()  #now the graph is align on edges 
```

```python
dsiris = pd.read_csv("iris.csv")
print(dsiris)
print(dsiris.head())
sns.set(style = 'whitegrid') # The "whitegrid" style adds a white background with grid lines to the plot,
ax = sns.stripplot(x='class' , y = 'sepal length' , data = dsiris) #one is data and another is class can take only one numeric data
plt.title("graph") # x data is class and y is seopal length and data is dsiris
plt.show()
```

```python
# conclusion ---> iris  sepal length is bigger in iris virginicia 
```

```python
sns.stripplot(x = 'class' ,y= 'sepal width' , data = dsiris)
plt.show()
```

```python
# sepal width of iris-selosa is more 
```

```python
sns.stripplot(x = 'sepal width' , y = 'class' , data = dsiris)
plt.title('horizontal')
plt.show()
```

```python
ax = sns.swarmplot(x = 'class' , y = 'sepal width' , data = dsiris)
plt.title('swarmplot')
plt.show()
```

```python
# note--> swarmplot is like a tree it shows better data representation of stripplot.
```
